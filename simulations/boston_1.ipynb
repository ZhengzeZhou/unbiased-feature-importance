```python
from sklearn.datasets import load_boston
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
import UFI
```

```python
import pandas as pd

boston = load_boston()
y = boston.target
X = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
X.shape
```

```python
X
```

```python
X["random"] = np.random.randn(X.shape[0])
```

```python
imp_name = X.columns
```

```python
# data = pd.DataFrame(X)
# data["y"] = y
# data.to_csv("boston.csv", index = False)
```

```python
All = []

for rep in range(100):
    
    regr = RandomForestRegressor(n_estimators = 100)
    regr.fit(X, y)
    
    All.append(regr.feature_importances_)

imp = np.average(All, axis = 0)

std = np.std(All, axis = 0)
```

```python
plt.figure(figsize=(18, 8))

pos = np.arange(len(imp_name))
bars = imp_name

barlist = plt.bar(pos, imp, yerr = std, ecolor = "red")
barlist[-1].set_color('black')
plt.xticks(pos, bars, fontsize = 15)

plt.ylabel("feature importance", fontsize = 18)

plt.savefig("SI_on_boston_data.png")
```
